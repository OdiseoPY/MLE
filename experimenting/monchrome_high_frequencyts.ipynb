### Read high_frequency separately:
(skip metadata rows, start at the readings table)

```python
hf_data = pd.read_csv("/content/Sample Input - Sample.csv.csv", skiprows = 20)
```

```python
hf_data.head()
```

#### Create timestamps:

```python
hf_data['timestamp'] = pd.to_datetime(hf_data['Date'] + ' ' + hf_data['Time']) + pd.to_timedelta(hf_data['Milli Sec'], unit='ms')
```

```python
hf_data.head()
```

```python
sensor_cols = [c for c in hf_data.columns if c not in ['Date','Time','Milli Sec', 'timestamp']]
len(sensor_cols)
```

```python
#melt for (tags per record)
hf_long = hf_data.melt(id_vars=['timestamp'], value_vars=sensor_cols, var_name='tag_name', value_name='tag_value')
```

```python
hf_long.head()
```

#### Melting

```python
hf_long.head()
```

```python
# Build mapping dicts
itemid_map = dict(zip(meta["ItemName"], meta["ItemId"]))
desc_map   = dict(zip(meta["ItemName"], meta["Comment"]))
unit_map   = dict(zip(meta["ItemName"], meta["Unit"]))
```

```python
hf_long["tag__id"]   = hf_long["tag_name"].map(itemid_map)
hf_long["tag__desc"] = hf_long["tag_name"].map(desc_map)
```

```python
hf_long['time_bucket']
```

```python
hf_long['time_bucket'] = hf_long['timestamp'].dt.floor('10s')
```

```python
hf_long.head()
```

```python
agg = (hf_long
       .groupby(['time_bucket', 'tag__id', 'tag_name', 'tag__desc', 'tag__unit'], as_index=False)
       .agg(tag__value=('tag_value', 'max')))
```

```python
agg
```

```python
agg.head()
```

```python
agg["tag__unit"] = 'mps' # sake of simplicity
```

```python
# Format timestamp
agg["event_timestamp"] = agg["time_bucket"].dt.strftime("%d/%m/%Y %H:%M:%S")

# Select final output columns in required order
out = agg[["event_timestamp", "tag__id", "tag_name", "tag__desc", "tag__value", "tag__unit"]]

display(out.head(20))
```
