## 1. Private Method Example
Single underscore (`_method`) is a convention meaning internal use.

Double underscore (`__method`) triggers name-mangling.

```python
class Database:
    def __init__(self):
        self._load_database()

    def _load_database(self):
        print("Loading DB...")

db = Database()
```

## 2. Magic Methods Example: `__init__` and `__add__`

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

p1 = Point(1,2)
p2 = Point(3,4)
p3 = p1 + p2
p3.x, p3.y
```
